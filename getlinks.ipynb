```python
new_df = pd.DataFrame({"Start":[],"End":[],"Dist":[]})
```

```python
for i in range(0,len(df)):
    if(df.iloc[i]["ID (Station ID)"]!=1):
        new_df.loc[len(new_df)] = {
            "Start":df.iloc[i-1]["Station Names"],
             "End":df.iloc[i]["Station Names"],
            "Dist":(df.iloc[i]["Dist. From First Station(km)"]-df.iloc[i-1]["Dist. From First Station(km)"])
            }
```
